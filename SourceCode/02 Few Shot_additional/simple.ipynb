```python
import torch
from transformers import OPTForCausalLM, AutoTokenizer
model = OPTForCausalLM.from_pretrained("facebook/opt-6.7b")
model =model.to("cuda")
tokenizer = AutoTokenizer.from_pretrained("facebook/opt-6.7b")
```

```python
import torch
from transformers import pipeline, set_seed
set_seed(1)
```

```python
generator=pipeline('text-generation', model="facebook/opt-6.7b", device=0)
```

```python
def has_numbers(inputString):
    return any(char.isdigit() for char in inputString)

def has_bracket(input_str):
    return "(" in input_str or "``" in input_str or ";" in input_str

import pandas as pd
pd.set_option('display.max_colwidth', None)

df = pd.read_csv(r"C:\Users\HADOOP\Desktop\NLP\train_prp copy.csv")
df.head()
arr=[]
for index, row in df.iterrows():
    tarr = row["sentence"].split("[Paraphrase]")
    arr.append((tarr[0].replace("[Sentence]:",""),tarr[1][1:]))

arr2=[]
for i in range(len(arr)):
    tup = arr[i]
    if(not has_numbers(arr[i][0]) and not has_numbers(arr[i][1])):
        if(not has_bracket(arr[i][0]) and not has_bracket(arr[i][1]) and len(arr[i][0]) < 70):
            arr2.append((tup[0].replace("\n","").replace("\r","").replace(" .","."), tup[1].replace("\n","").replace(" .",".")))
arr2
import random
shuff = random.shuffle(arr2)
text=""
for i in range(20):
    text = text+"text: "+arr2[i][0]+"\n"+"paraphrase: "+arr2[i][1]+"\n\n\n"
text = text[:-1]
print(text)
```

```python
prompt = f"""text: The abbey is registered as a regional cultural asset.
paraphrase: The abbey is registered as a regional asset of cultural importance.


text: occupied Finland , Prussia and Poland.
paraphrase: Poland , East Prussia and occupied Finland.


text: Sammamish lake enters the Issaquah Creek in the park.
paraphrase: Lake Sammamish enters Issaquah Creek in the park.


text: Some have negative and some positive effects.
paraphrase: Some have positive and some negative effects.


text: The top ten was completed by Zhou , Sette Câmara and Sekiguchi.
paraphrase: Zhou , Sette Câmara and Sekiguchi completed the top ten.


text: Dye rode in Mauritius after eight years in Hong Kong.
paraphrase: Dye rode in Hong Kong in Mauritius after eight years.


text: NLP is difficult to learn.
paraphrase:"""


input_ids = tokenizer(prompt, return_tensors="pt").input_ids.to("cuda")
generated_ids = model.generate(input_ids, do_sample=True, temperature=0.9, max_length=2100, remove_invalid_values=True)
generated_text = tokenizer.decode(generated_ids[0])
print(generated_text)

x = input_ids.cuda()
del x
torch.cuda.empty_cache()
```
